```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import re
import os
import numpy as np
import shutil
import re
import pandas as pd
```

```python
dirs = "./html"
if os.path.exists(dirs):
    shutil.rmtree(dirs)
os.makedirs(dirs)
urls = []
for year in range(2004, 2020):#2014,2020
    url = "https://www.nj.gov/labor/lwdhome/warn/"+str(year)+"/warnindex.html"
    page = requests.get(url)
    soup = BeautifulSoup(page.text, 'html.parser')
    #div_data = soup.find("p")#soup.find("div", class_="genericContent")
    div_data = soup.select('div[class*="contentArea"]')[0]
    table_data = div_data.findAll('a')[3:]
    urls += table_data
#urls
```

```python
company = []
city = []
effectivedate = []
workforceaffected = []

for ele in urls:
    sub_url = ele.get('href')
    html_name = "_".join(ele.text.split()[:2])
    if "http:" not in sub_url:
        sub_url = "https://www.nj.gov"+ele.get('href')
    sub_page = requests.get(sub_url)
    with open("./html/"+html_name+".html", 'w',encoding = 'utf8') as file: 
        file.write(sub_page.text)
    sub_soup = BeautifulSoup(sub_page.text, 'html.parser')

    #sub_soup2 = sub_soup.find("div", class_="genericContent")
    txt = []
    for ele in sub_soup.findAll("td"):
        txt.append(ele.text)
    regex = re.compile(r'[\n\r\t\xa0]')
    txt =[regex.sub(" ", s).strip() for s in txt]

    for i, v in enumerate(txt[4:]):    
        remainder = i % 4
        if remainder == 0:
            company.append(v)
        if remainder == 1:
            city.append(v)
        if remainder == 2:
            effectivedate.append(v)
        if remainder == 3:
            workforceaffected.append(v)
    print(html_name+"is done!")
company = [ls.replace('Â', '') for ls in company]
city = [ls.replace('Â', '') for ls in city]
effectivedate = [ls.replace('Â', '') for ls in effectivedate]
workforceaffected = [ls.replace('Â', '') for ls in workforceaffected]
d1 = {'company': company, 'city': city, 'effectivedate': effectivedate, 'workforceaffected': workforceaffected}
df = pd.DataFrame(data=d1)
```

```python
df.to_csv("final_result.csv", index = False)
```
